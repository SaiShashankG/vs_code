```python
def mirror(n):
    s = str(n)
    mirror_str = s[::-1]
    return int(mirror_str)

def find_min_mirror_distance(nums): 
    min_distance=nums[0]
    for i, num1 in enumerate(nums):
        mirrored = mirror(num1)
    
        for j in range(i + 1, len(nums)):
            if nums[j] == mirrored:
                distance = j - i
                min_distance = min(min_distance, distance)
    
    return min_distance if min_distance != float('inf') else -1

nums=[9,9]
print(find_min_mirror_distance(nums))
```

```python
def bubble_sort(arr):
    n=len(arr)
    for i in range(n):
        isSwap=False
        for j in range(n-i-1):
            if arr[j]>arr[j+1]:
                isSwap=True
                arr[j],arr[j+1]=arr[j+1],arr[j]
        if not isSwap:
            break
    return arr
```

```python
def insertion_sort(arr):
    n=len(arr)
    for i in range(1,n):
        key=arr[i]
        j=i-1
        # Alternate Approach
        # while j>=0 and arr[j]>key:
        #     arr[j],arr[j+1]=arr[j+1],arr[j]
        #     j-=1
        while j>=0 and arr[j]>key:
            arr[j+1]=arr[j]
            j-=1
        arr[j+1]=key
    return arr
```

An input string is valid if:

Open brackets must be closed by the same type of brackets.
Open brackets must be closed in the correct order.
Every close bracket has a corresponding open bracket of the same type. `LeetCode`

```python
def valid_parenthesis(s):
    n=len(s)
    if n&1==1: ##Length odd case
        return False
    st=[]
    for ch in s:
        if ch=="(" or ch=="[" or ch=="{":
            st.append(ch)
        else:
            if len(st)==0:
                return False
            top=st.pop()
            if ch==")" and top!="(":
                return False
            elif ch=="]" and top!="[":
                return False
            elif ch=="}" and top!="{":
                return False
    return len(st)==0
```
